```python
#Un abb es como un arbol genialofico
#pero con NUMEROS
#1. Cada uno tiene como maximo 2 hijos 
#2. Izquierda: Numeros menores 
#3. Derecha: Numeros mayores


#Leccion 1 : Creando nodo
#Un nodo en como una caja 
#Guarda: el valor (numero)
#Guarda: 2 flechas apuntando asus hijos

class Nodo:
    def __init__(self,valor):
        self.valor= valor
        self.izq= None
        self.der= None

raiz = Nodo(10)
print(f"Nodo creado con valor: {raiz.valor}")
print(f"Hijo izquierdo: {raiz.izq}")
print(f"Hijo derecho: {raiz.der}")
```

```python
#Crea un Nodo con valor 50 y muestra su valor

raiz02= Nodo(50)

print(f"Valor del nodo es {raiz02.valor}")
```

```python
#Conectando nodos
# Crea un arbol peque√±o 

raiz03= Nodo(10)
raiz03.izq =Nodo(5) #Va a la izqierda por ser menor de 10 (Osea la raiz)
raiz03.der= Nodo(15) #Va a la derecha por que es mayor que 10 (Osea la raiz)


print(f"Valor de la raiz: {raiz03.valor}")
print(f"Hijo izquierod tiene el valor: {raiz03.izq.valor}")
print(f"Hijo derecho tiene el valor: {raiz03.der.valor}")
```

```python
#III: Insertar valores:
# Regla:
# Si el valor es menor ira a la izquierda 
# Si el valor es mayor ira a la derecha
# Repetir hasta encontrar un espacio vacio

def insertar(raiz,valor):
    #Si arbol esta vacio, Crea la raiz
    if raiz is None:
        return Nodo(valor)
    #si el valor es menor que la raiz ira a la izquierda
    if valor < raiz.valor:
        raiz.izq= insertar(raiz.izq, valor)
    #si el valor es mayor a la raiz ira a la derecha
    else:
        raiz.der = insertar(raiz.der, valor)
    return raiz


# Ejemplo: 

raiz05 = None

raiz05 = insertar(raiz05, 10)
raiz05 = insertar(raiz05, 5)
raiz05 = insertar(raiz05, 15)
raiz05 = insertar(raiz05, 3)


print(f"La raiz es: {raiz05.valor}")
print(f"Hijo izquierdo: {raiz05.izq.valor}")
print(f"Hijo derecho: {raiz05.der.valor}")

```
